```python
# from data i am dropping the coloum 'anmeOrig' and 'nameDest' as i wont use them in my model training

df_model = df_cc.drop(['nameOrig','nameDest'],axis=1,inplace=False)
df_model
```

### Labelencoding for categorical features



```python
#here i am going to do labelencoding to the type feature

from sklearn.preprocessing import LabelEncoder

lab = LabelEncoder()
df_model['type']=lab.fit_transform(df_model['type'])
```

```python
df_model
```

### modelelling using ml alogs


```python
X=df_model.drop('isFraud',axis = 1)
y=df_model['isFraud']

X
```

```python
y
```

```python
#lets split the data for train and test data

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

### LogisticRegression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lg_reg = LogisticRegression()
lg_reg.fit(X_train,y_train)

lg_pred = lg_reg.predict(X_test)

print('the prediction has an acc_score as : {:0.5f}'.format(accuracy_score(y_test,lg_pred)))
```
